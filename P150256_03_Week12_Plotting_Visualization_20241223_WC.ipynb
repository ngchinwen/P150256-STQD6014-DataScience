### Histogram


*   **`bins`**: Bins are the intervals into which the data is divided for counting and display in the histogram.
    - Having 20 bins means the data range will be split into 20 intervals
*   **`alpha`**: controls the transparency of the bars.
    - A value of 1.0 would be completely opaque, while 0.0 would be completely transparent.



```python
# Histogram - plot 1
rng = np.random.RandomState(12345) #setting the seed number

#getting 100 values in a normal disbribution that is mean =0 and sd = 1
#chopping it up to 20 bins
# this will come out in final exam
plt.hist(rng.standard_normal(100),bins=8,color="pink",alpha=0.5);
```

### **Place all the subplots into one panel**

1.   **plt.figure()** - create a new figure
2.   **add_subplot** - to create one or more subplots
3.   **Subplots** are numbered sequentially **starting from 1, going from left to right and top to bottom.**
    - a **(2, 2, 1)** argument means we're adding the first subplot (top-left corner) in the 2x2 grid.
    2,2 is the dimensions of the grid. 1 is the position in that 2x2 plot.
  


```python
# Create one or more subplots
# 2 x 2; 4 plots in total
# Plot axis
rng = np.random.RandomState(12345)
fig = plt.figure(dpi=300)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax1.hist(rng.standard_normal(100), bins=20, color='black', alpha=0.3);
ax2.scatter(np.arange(30), np.arange(30) + 3 * rng.standard_normal(30));
ax3.plot(rng.standard_normal(50).cumsum(), color='black', linestyle='dashed');

# To save the figures
#plt.savefig('/content/drive/MyDrive/STQD6014_Executive/_SEM1_20242025/Figures/combi3.pdf')
plt.savefig('/content/drive/MyDrive/Data Science/STQD6014/20241224_w12_combi3.pdf')
```

### **Adjusting the spacing around subplots**

```python
# An example that shrink the spacing of the subplots to zero
# Change the value of wspace and hspace
rng = np.random.RandomState(12345)
 #sharex means all the plots shares the same x-axis
 #sharey makes the plot have a y axis for each subplot
fig,axes=plt.subplots(2,2,sharex=True,sharey=False)

# Using a nested loop to populate 2x2 grid
# for range, 0,1, so it will be 0,0 - 0,1 - 1,0 - 1,1
for i in range(2):
    for j in range(2):
        axes[i, j].hist(rng.standard_normal(500), bins=50,
                        color='black', alpha=0.5)

# Adjusting the spacing
fig.subplots_adjust(wspace=0.8, hspace=-0.2)
```

## Drawstyles
*   `default`: makes a straight between each data point
* **Step-like plots**
<br> the only thing that changes is how the horizontal line is treated
** `steps-pre`: horizontal line is inline with the previous data point's y-value until it reaches the next data point at which it will step up
** `steps-mid`:
** `steps-post`:


```python
# Drawstyle default
rng = np.random.RandomState(12345)
dat = rng.standard_normal(50).cumsum()
plt.plot(dat, color='red', linestyle='dashed', label='Red ^_^',drawstyle='default')
plt.legend();
```

```python
# Drawstyle steps-mid
# step-like graph is good for data that changes in a step-like manner
# example: daily stock prices, signal levels, cumulative distribution
plt.plot(dat, color='blue', linestyle='dashed',
         drawstyle='steps-mid', label='Blue ^_^');
plt.legend();
```

```python
# Combine
rng=np.random.RandomState(12345)
data=rng.standard_normal(5).cumsum()
plt.plot(data, color='black', linestyle='dashed', label='Default')
plt.plot(data, color='red', linestyle='dashed',
         drawstyle='steps-pre', label='steps_pre',alpha=0.4)
plt.plot(data, color='green', linestyle=':',
         drawstyle='steps-mid', label='steps_mid',alpha=0.4)
plt.plot(data, color='blue', linestyle=':',
         drawstyle='steps-post', label='steps_post',alpha=0.4)
plt.legend();
#plt.savefig('/content/drive/MyDrive/lines.pdf', dpi=300)
plt.savefig('/content/drive/MyDrive/Data Science/STQD6014/20241224_w12_lines.pdf', dpi=30)
```

### ***Setting the title, axis labels, ticks, and tick labels***

```python
# An example
rng = np.random.RandomState(12345)
fig, ax = plt.subplots()
ax.plot(rng.standard_normal(1000).cumsum());

# Changex x-axis ticks annotation
ax.set_xticks([0,50,500,750,1000]) #can change where to annotate

# Cabel the x-axis ticks with names
ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize=18)
ax.set_xlabel('Stages')
ax.set_ylabel('Values')
ax.set_title('My First Graph :3c');
```

### **Annotations and Drawing on a Subplot**

```python
# The SPX data
from datetime import datetime

fig, ax = plt.subplots()

#reads this link: https://github.com/wesm/pydata-book/blob/3rd-edition/examples/spx.csv
data = pd.read_csv('https://bit.ly/3iofIY8', index_col=0, parse_dates=True)
#print(data.columns) #only has one column spx

spx = data['SPX']

spx.plot(ax=ax, color='blue')
#print(spx)

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 60),
                xytext=(date, spx.asof(date) + 240),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/10/2007', '1/1/2009'])
ax.set_ylim([0, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis');
```

### **Adding shapes on the same plot (IMPORTANT)**
- Will appear in exam (15 marks)

```python
# Adding shapes to the same plot
fig, ax = plt.subplots()

#create a rectangle starting from (x,y) of .2,.75
#then from that point
#expand from x of .2 + 0.4 = 0.6
#expand from y of .75 + 0.15 = 0.9
rect=plt.Rectangle((0.2,0.75),0.4,0.15,color='black',alpha=0.3)

#create a circle starting from (x,y) point of .7,.2 #then from that point
#expand the radius to .15 from that xy point of .7,.2 so it's a total diameter of .3

#x side is .7 +- .15 #left will be .55   #right will be .85
#y side is .2 +- .15 #bottom touches .05 #top will be .35
circ=plt.Circle((0.7,0.2),0.2,color='blue',alpha=0.3)

#each coordinate is a point in the polygon shape
pgon=plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='green',alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)

#starting from a x,y point of .2,. # make a pentagon(5) # with radius 0.1
#the top touches 0.8
#the bottom should touch 0.6 if it was a hexagon
pentagon=patches.RegularPolygon((0.2,0.7),6,radius=0.1,color='red')
ax.add_patch(pentagon)

#starting from the xy
#make the x expand to 0.2 on both sides
#make the y expand upwards/downwards by 0.1
ellipse=patches.Ellipse((0.5,0.5),0.2,0.1,color='yellow')
ax.add_patch(ellipse)

square=patches.Rectangle((0.3,0.1),0.2,0.2,color='black',alpha=0.3)
ax.add_patch(square);
```

```python
# Challenge: plot a pentagon, ellipse, square figures in the exact same figure
fig, ax = plt.subplots()

# Pentagon
  # Center, sides, radius
pentagon=patches.RegularPolygon((0.2,0.7),6,radius=0.1,color='red')
ax.add_patch(pentagon)

# Ellipse
  # Center, width, height
ellipse=patches.Ellipse((0.5,0.5),0.2,0.1,color='yellow')
ax.add_patch(ellipse)

# Square
  # Lower-left, width, height
square=patches.Rectangle((0.3,0.1),0.2,0.2,color='green',alpha=0.3)
ax.add_patch(square);

plt.show()  # Display the figure
```

```python
# Adding shapes to a plot
import matplotlib.patches as mpatches
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='black', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='blue', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='green', alpha=0.5)
sqr = plt.Rectangle((0.3, 0.1), 0.2, 0.2, color='black', alpha=0.3)

# Add ellipse
ellipse = mpatches.Ellipse((0.5,0.5), 0.2, 0.1)

# Add regular hex
hex = mpatches.RegularPolygon((0.7,0.8), 6, radius=0.1)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
ax.add_patch(sqr)
ax.add_patch(ellipse)
ax.add_patch(hex);
```
